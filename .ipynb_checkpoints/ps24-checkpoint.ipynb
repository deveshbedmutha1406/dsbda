```python
not_affected = df[df['area'] == 0]
not_affected
```

```python
partially = df[(df['area'] <= df['area'].mean()) & (df['area'] > 0)]
```

```python
partially
```

```python
major = df[df['area'] > df['area'].mean()]
major
```

```python
# merge
```

```python
merge = pd.concat([partially, major])
merge
```

```python
merge.reset_index(drop=True, inplace=True)
merge.head(10)
```

```python
df2=df
df2.head()
```

```python
for i in df2['area'].values:
    if i == 0.0:
        df2['area'] = df2['area'].replace(i, 'Not Affected')
    elif i > 0.0 and i <= 250.0:
        df2['area'] = df2['area'].replace(i, 'partially affected')
    else:
        df2['area'] = df2['area'].replace(i, 'mostly affected')
df2.head()
```
