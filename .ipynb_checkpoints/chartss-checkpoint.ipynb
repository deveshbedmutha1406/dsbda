```python
import pandas as pd
import numpy as np
df = pd.read_csv('DSBDALExam DataSets/heart.csv')
df.head()
```

```python
# histogram
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.title("age distribution using hist plot")
sns.set_style("darkgrid")
sns.histplot(x='age', data=df, hue='output')
```

```python
plt.figure(figsize=(10,10))
plt.title("scatter plot")
sns.scatterplot(x='age', y='chol', data=df, hue='sex', marker='o')
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='age', y='chol', data=df)
```

```python
sns.lineplot(x='age', y='chol', data=df)
```

```python
sns.boxplot(x='age', y='chol',data=df)
```
