```python
def outlier(df2, drop=False):
    for each_feature in df2.columns:
        var1 = df2[each_feature]
        IQ1 = np.percentile(var1, 25.)
        IQ2 = np.percentile(var1, 75.)
        IRQ = IQ2 - IQ1
        threshold = IRQ * 1.5
        l1 = var1[~((var1 >= IQ1 - IRQ) & (var1 <= IRQ + IQ2))].index.tolist()
        
        if drop == False:
            print(f"{each_feature} contains {len(l1)} outliers")
        else:
            df2.drop(l1, inplace=True, errors="ignore")

cont_features = df.loc[:,['age', 'fnlwgt', 'capital-gain', 'capital-loss', 'hours-per-week', 'education-num']]
outlier(cont_features)
cont_features
```

```python
from sklearn.preprocessing import LabelEncoder
obj = LabelEncoder()
df['workclass'] = obj.fit_transform(df['workclass'])
df['workclass'].value_counts()
```

```python
l2 = ['education-num', 'maritial-status', 'occupation', 'relationship','race', 'sex', 'native-country', 'salary']
```

```python
for ele in l2:
    df[ele] = obj.fit_transform(df[ele])
```

```python
df.dtypes
```

```python
df['education'] = obj.fit_transform(df['education'])
```

```python
df.dtypes
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
X = df.drop(['salary'], axis=1).values
y = df['salary']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_text, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)
```

```python
model.fit(x_train, y_train)
y_pred = model.predict(x_text)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred) * 100
```

```python
metrics.precision_score(y_test, y_pred) * 100
```

```python
metrics.recall_score(y_test, y_pred) * 100
```

```python
metrics.f1_score(y_test, y_pred)
```
